## Feature Selection: Duplicated Features

**Duplicate Features** are the ones when both the features in essenece are identical. Duplicate features might be already present in the dataset or might be introduced using one-hot encoding.

```python
# Import Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Load Dataset
df = pd.read_csv('./dataset/Santander-Customer-Satisfaction-data/train.csv', nrows=15000)
df.head()
```

```python
X = df.drop(['TARGET'], axis=1)
X.head()
```

```python
y = df['TARGET']
y.head()
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
data_t = X_train.T
data_t.head()
```

```python
# Search for Duplicated Columns
data_t.duplicated().sum()
```

```python
# Visualize Duplicated Rows
data_t[data_t.duplicated()]
```

```python
# Duplicated Features
duplicated_features = data_t[data_t.duplicated()].index.values
```

```python
duplicated_features
```

```python
#  Drop Duplicated Features
unique_df = data_t.drop_duplicates(keep='first').T
unique_df.shape
```

This shows that, we started from a dataset containing the duplicated features and hence the total number of columns 371 which after the duplicate features removal has come down to 271.

```python
# Columns Removed from original Dataframe
removed_features = [col for col in df.columns if col not in unique_df.columns]
removed_features
```

### Remove Duplicate Features from Large Datasets

```python
df = pd.read_csv('./dataset/Santander-Customer-Satisfaction-data/train.csv', nrows=50000)
df.head()
```

```python
X = df.drop(['TARGET'], axis=1)
X.head()
```

```python
y = df['TARGET']
y.head()
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# check for duplicated features in the training set
duplicated_feat = []
for i in range(0, len(X_train.columns)):
    if i % 10 == 0:  # this helps me understand how the loop is going
        print(i)
 
    col_1 = X_train.columns[i]
 
    for col_2 in X_train.columns[i + 1:]:
        if X_train[col_1].equals(X_train[col_2]):
            duplicated_feat.append(col_2)
            
len(duplicated_feat)
```

```python
# Drop Duplicate Features from Training and Test Set
X_train.drop(labels=duplicated_feat, axis=1, inplace=True)
X_test.drop(labels=duplicated_feat, axis=1, inplace=True)
 
X_train.shape, X_test.shape
```
